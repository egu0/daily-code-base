```python
import os
from dotenv import load_dotenv

load_dotenv()

import aisuite as ai

# Define the client. You can use this variable inside your graded functions!
CLIENT = ai.Client(
    provider_configs={"deepseek": {"api_key": os.getenv("DEEPSEEK_API_KEY")}}
)
```

```python
def generate_draft(topic: str, model: str = "deepseek:deepseek-v4-flash") -> str:

    prompt = f"""
    Write a well-structured draft essay in response to the following topic.
    The draft should include an introduction, body, and conclusion.

    topic:
    {topic}
    """

    # Get a response from the LLM by creating a chat with the client.
    response = CLIENT.chat.completions.create(
        model=model,
        messages=[{"role": "user", "content": prompt}],
        temperature=1.0,
    )

    return response.choices[0].message.content
```

```python
user_question = "Should people be encouraged to eat less meat?"
draft_v1 = generate_draft(user_question)
print(draft_v1)
```

```python
def reflect_on_draft(draft: str, model: str = "deepseek:deepseek-v4-flash") -> str:

    prompt = f"""
    Reflect critically on the following draft essay.
    Identify areas for improvement in structure, clarity, argument strength, or style.
    Provide constructive feedback in paragraph form.

    Draft:
    {draft}
    """

    # Get a response from the LLM by creating a chat with the client.
    response = CLIENT.chat.completions.create(
        model=model,
        messages=[{"role": "user", "content": prompt}],
        temperature=1.0,
    )

    return response.choices[0].message.content
```

```python
feedback = reflect_on_draft(draft_v1)
print(feedback)
```

```python
def revise_draft(original_draft: str, reflection: str, model: str = "deepseek:deepseek-v4-flash") -> str:

    prompt = f"""
    Revise the following essay draft using the feedback provided.
    Make improvements in clarity, coherence, argumentation, and flow.
    Return only the improved essay.

    Original Draft:
    {original_draft}

    Feedback:
    {reflection}
    """

    # Get a response from the LLM by creating a chat with the client.
    response = CLIENT.chat.completions.create(
        model=model,
        messages=[{"role": "user", "content": prompt}],
        temperature=1.0,
    )

    return response.choices[0].message.content
```

```python
draft_v2 = revise_draft(draft_v1, feedback)
print(draft_v2)
```
