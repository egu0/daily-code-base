## 数据加载与持久化

### csv 格式

````python
df = pd.DataFrame(data = np.random.randint(0, 50, size = (50, 5)),
               columns = ['IT', '化工', '生物', '教师', '土木'])
df.head()
````

````python
# 持久化到 csv 文件
#   sep: 分隔符，默认为英文逗号
#   header: 是否保存列索引
#   index: 是否保存行索引。文件被加载时，默认的行索引会被作为一列
df.to_csv('salary.csv', sep=':', header=True, index=True)
````

salary.csv 前几行数据：
```
:IT:化工:生物:教师:土木
0:30:38:23:48:10
1:10:31:3:37:11
2:16:41:28:26:42
3:36:39:1:47:34
4:14:38:8:2:7
```

````python
# 读取 csv 文件
#   sep: 指定分隔符
#   header: 指定列索引
#   index_col: 指定行索引
df2 = pd.read_csv('salary.csv', sep=':', header=[0], index_col=0)
df2.head()
````

**赋值操作**

````python
#一维序列，表示 10 个学生的 Pytorch 成绩
s = pd.Series(np.random.randint(0, 150, size=10),
              index=list('ABCDEFGHIJ'),
              name='Pytorch')

#二维数组，表示学生成绩
df = pd.DataFrame(np.random.randint(0, 150, size=(10, 3)),
                  index=list('ABCDEFGHIJ'),
                  columns=['Pandas', 'Tensorflow', 'Keras'])
df.head()
````

````python
df['Java'] = np.random.randint(0, 100, size = 10)
df.head()
````

````python
#增加一列，会根据行索引自动对齐
df['Pytorch'] = s
df.head()
````

````python
#按标签赋值
df.loc['A', 'Pytorch'] = 150

#按位置赋值
df.iloc[1, 2] = 150

df.head()
````

````python
#对某一列进行赋值
#df.loc[:, 'Pandas'] = 10
df.loc[:, 'Pandas'] = np.array([100] * 10)
df.head()
````

````python
#对某一行进行赋值
#df.iloc[0] = 92
df.iloc[0] = np.array([92] * 5)
df.head()
````

````python
cond = df < 92
df[cond] = 92
df.head()
````

2

使用字典创建 DataFrame，行索引为 a\~z，列索引是身高（160\~185）、体重（50\~90）、学历（无、本科、硕士、博士）
身高和体重使用 Numpy 随机生成，学历数据先创建数组 `edu=np.array(["无","本科","硕士","博士"])`，然后使用花式索引从四个数据中选择 26 个数据

````python
edu = np.array(["无","本科","硕士","博士"])
random_idx = np.random.randint(0, 4, size = 26)
edu[random_idx]
````

````python
import string

df = pd.DataFrame(data = {
                    'height': np.random.randint(160, 186, size=26),
                    'weight': np.random.randint(50, 91, size=26),
                    'education': edu[random_idx],
                    }, 
                  index=[letter for letter in string.ascii_lowercase])
df.head()
````
