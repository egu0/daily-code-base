### No.6

创建一个每一行都是从 0 到 4 的 5*5 矩阵

```python
arr = np.zeros(shape = (5,5), dtype = np.int16)

for i in range(1, 5):
    arr[:, i] = arr[:, i-1] + 1

arr
```

### No.10

如何根据第 3 列大小顺序来对一个 5*5 矩阵排序？（提示：查阅 argsort 方法；根据某一列重新组织行顺序）

```python
arr = np.random.randint(0, 10, size=(5,5))
arr
```

```python
sort_idx = arr[:, 2].argsort()
sort_idx
```

```python
# 按照第三列的大小顺序重新排列所有行
arr[sort_idx]
```
