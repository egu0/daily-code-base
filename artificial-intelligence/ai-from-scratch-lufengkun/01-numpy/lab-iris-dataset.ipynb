## Lab

读取 【iris 数据集】（`iris_dataset.csv`） 中的花萼长度（第一列）数据，对其进行排序、去重、求和、累积和、均值、方差、标准差、最大值、最小值。

```python
import numpy as np
```

```python
data = np.loadtxt('iris_dataset.csv', delimiter=',')
data.shape
```

```python
col1 = data[:, 0]
```

```python
col1
```

```python
#排序
col1[col1.argsort()]
```

```python
#去重
np.unique(col1)
```

```python
#求和
np.sum(col1)
```

```python
#累积和
col1.cumsum()
```

```python
#均值
np.mean(col1)
```

```python
#最大值
col1.max()
```

```python
#最小值
col1.min()
```

```python
#方差
col1.var()
```

```python
#标准差
col1.std()
```
