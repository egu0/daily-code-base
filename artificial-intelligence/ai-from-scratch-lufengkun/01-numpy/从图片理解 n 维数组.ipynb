````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
img = plt.imread('cat.png')
img.shape
````

````python
img.ndim
````

````python
plt.imshow(img)
````

````python
# 去除阿尔法通道
rgb_img = img[:, :, :3]
plt.imshow(rgb_img)
````

````python
# 红绿蓝 -> 蓝绿红
plt.imshow(rgb_img[:, :, ::-1])
````

````python
# 红绿蓝 -> 绿蓝红
plt.imshow(rgb_img[:, :, [1, 2, 0]])
````
